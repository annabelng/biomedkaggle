```python
import re
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
import os
%matplotlib inline

# some code to control figure aeshetics
#sns.set_style("ticks")
#sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
basedir = os.getcwd()
datadir = basedir + '/data'
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = pd.read_csv(datadir + '/test_label.csv')
y_true = y_true.iloc[:,0].tolist()

y_pred = pd.read_csv(basedir + '/predictions')
y_pred = y_pred.iloc[:,0].tolist()

cm = confusion_matrix(y_true, y_pred, labels=[0,1,2,3,4,5,6,7,8])
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = [0,1,2,3,4,5,6,7,8])
disp.plot()
plt.show()
```
